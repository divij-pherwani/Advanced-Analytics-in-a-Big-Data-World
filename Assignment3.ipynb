# Construction of predictive model that can predict the channel based on a chat message

```python
# Loading CSV file as a spark dataframe
df_spark = spark.read.option("header",True).csv("twitch-data2.csv")
#spark.read.csv("twitch-data.csv").option("header","true")
```

```python
# Preprocessing - Dropping duplicates
df_spark = df_spark.dropDuplicates()
```

```python
# Preprocessing -cleaning channel names if messy
df_spark = df_spark.where((df_spark.channel=='asmongold') | (df_spark.channel=='loltyler1'))
```

```python
# Checking channel distribution
df_spark.groupby('channel').count().show()
```

```python
# Text Mining Analysis

# Tokenization
tokenizer = Tokenizer(inputCol='content',outputCol='mytokens')

# Stopword removal
stopwords_remover = StopWordsRemover(inputCol='mytokens',outputCol='filtered_tokens')

# Vectorization
vectorizer = CountVectorizer(inputCol='filtered_tokens',outputCol='rawFeatures')

# Inverse Data Frequency (idf)
idf = IDF(inputCol='rawFeatures',outputCol='vectorizedFeatures')
```

```python
# Label encoding for channel names 
labelEncoder = StringIndexer(inputCol='channel',outputCol='label').fit(df_spark)
labelEncoder.transform(df_spark).show(5)
```

```python
# Label encoding the dataframe
df_spark = labelEncoder.transform(df_spark)
```

```python
# Splitting the dataframe in training and testing sets
(trainDF,testDF) = df_spark.randomSplit((0.7,0.3),seed=42)
```

## Logistic Regression

```python
# Logistic regression
lr = LogisticRegression(featuresCol='vectorizedFeatures',labelCol='label')
```

```python
# Setting pipeline
pipeline = Pipeline(stages=[tokenizer,stopwords_remover,vectorizer,idf,lr])
```

```python
# Fitting model 
lr_model = pipeline.fit(trainDF)
# Predicting test set
predictions = lr_model.transform(testDF)
# Displaying predictions 
predictions.show()
```

```python
# Showing selected columns of prediction dataframe
predictions.select('rawPrediction','probability','channel','label','prediction').show(10)
```

```python
# Evaluating the accuracy
evaluator = MulticlassClassificationEvaluator(labelCol='label',predictionCol='prediction',metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
accuracy
```

## Gradient Boosting Tree

```python
# Declaring the GBT classifier
gbt = GBTClassifier(labelCol="label", featuresCol="vectorizedFeatures", maxIter=10)
# Setting pipeline
pipeline = Pipeline(stages=[tokenizer,stopwords_remover,vectorizer,idf,gbt])
# Fitting model
gbt_model = pipeline.fit(trainDF)
# Predicting test set
predictions = gbt_model.transform(testDF)
predictions.show()
```

```python
# Filtering prediction columns
predictions.select('rawPrediction','probability','channel','label','prediction').show(10)
```

```python
# Evaluating accuracy
evaluator = MulticlassClassificationEvaluator(labelCol='label',predictionCol='prediction',metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
accuracy
```

## Linear Support Vector Machine

```python
# Initializing Linear SVC model
lsvc = LinearSVC(maxIter=10, regParam=0.1, labelCol="label", featuresCol="vectorizedFeatures")
# Setting pipeline 
pipeline = Pipeline(stages=[tokenizer,stopwords_remover,vectorizer,idf,lsvc])
# Fitting model
lsvc_model = pipeline.fit(trainDF)
# Predicting test set
predictions = lsvc_model.transform(testDF)
predictions.show()
```

```python
# Filtering prediction columns 
predictions.select('rawPrediction','channel','label','prediction').show(10)
```

```python
# Evaluating accuracy
evaluator = MulticlassClassificationEvaluator(labelCol='label',predictionCol='prediction',metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
accuracy
```

## Showing that our model can make predictions in a "deployed" setting on new messages

```python
# Label dictionary
label_dict = {'loltyler1':1.0,'asmongold':0.0}
label_dict
```

```python
# Loading saved model
model = PipelineModel.load("twitch.model")
model
```

```python
# counter for live data prediction
counter = 1

# correct prediction counter
correct = 0

# wrong prediction counter
wrong = 0


class Bot(commands.Bot):    
    
    def __init__(self):
        # intializing bot to stream selected channels 
        super().__init__(token, client_id= cid,prefix=['!', '?'],nick = 'divij0476', initial_channels=['loltyler1', 'asmongold'])
    
    async def event_message(self, message):
        global model
        global label_dict
        global pdf
        global counter
        global correct
        global wrong 

        # fetching message content  
        content = message.content
        
        # fetching author name
        if message.author:
            author = message.author.name
        else:
            author = "Unknown"        
        
        # fetching message channel name
        channel = message.channel.name
        
        # Setting labels 
        
        label = 0.0 
        if channel == "loltyler1":
            label = label_dict['loltyler1']
        elif channel == "asmongold":
            label = label_dict['asmongold']
        
        # Declaring empty dictionary
        temp = {'author' : str(author), 'content' : str(content), 'channel': str(channel), 'label': label}
        
        # converting dictionary to dataframe pandas 
        df = pd.DataFrame([temp])
        
        # converting pandas dataframe to spark
        spark_df = spark.createDataFrame(df) 
        
        # predicting values using model for single spark dataframe
        prediction = model.transform(spark_df)
        
        # Filtering prediction columns
        pred = prediction.select('author', 'content', 'channel', 'label', 'prediction')
        
        # predicted label value 
        l = pred.collect()[0].prediction
             
        # predicted vs actual value for counter 
        if int(l) == int(label):
            correct += 1
        else:
            wrong += 1
            
        
        # Printing results 
        print("######### " + str(counter) + " ##########")
        pred.show()
        print("Correct count: {} & Wrong count: {}".format(correct, wrong))
        counter += 1
        
        await self.handle_commands(message)
        
try:
    bot = Bot()
    bot.run()
except RuntimeError:
     pass
```
